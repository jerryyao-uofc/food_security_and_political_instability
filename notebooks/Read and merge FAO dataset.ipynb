```python
# Manipulating One Dataframe

def dataframe_transform_item_code(df, dict_meaning, i):
    '''
    transform dataframe that has "item code" and "element code " specifically
    
    
    return: 
        1. an mapping from identifier to meaning
    '''
    
    if 'Item Code' not in df.columns.tolist():
        raise ValueError('The inputted dataframe does not have item-code')
    
#     df['randNumCol'] = np.random.randint(1, 50000, df.shape[0])       
    df['identifier'] = (df['Item Code'].astype(str) + df['Element Code'].astype(str) ).astype(int)
    
    for each in df['identifier'].unique():
        

        item_row = (df[df['identifier'] == each]).iloc[0]
        item_name =str(item_row['Item'])+ " " + str(item_row['Element'])
        
#         if each == 210326125:
#             print("hello", i)
        
#         if each == '210326125':
#             print("hello", i )
        
        dict_meaning[each] = item_name
    
    
    if 'Area Code' in df.columns: 
        ac = 'Area Code'
    else: 
        ac = 'Reporter Country Code'
        
    df_step = df.filter([ac, 'Year', 'identifier', 'Value'])

    if i == 17:
        # some prep work before pivot
        df_step.Year = df_step.Year.str[-4:]
        df_step['Value'] = df_step['Value'].str.replace('<', '')
        df_step.Value = pd.to_numeric(df_step.Value)

     
        df_step['Year'] = df_step['Year'].astype(int)
    
    df_final = df_step.pivot_table('Value', ['Year', ac], 'identifier')


    

    
    return df_final
```

```python
def dataframe_transform_sex_code(df, dict_meaning, i ):
    '''
    transform dataframe that has "item code" and "element code " specifically
    
    
    return: 
        1. an mapping from identifier to meaning
    '''
    
    if 'Sex Code' not in df.columns.tolist():
        raise ValueError('The inputted dataframe does not have item-code')
    
#     df['randNumCol'] = np.random.randint(1, 50000, df.shape[0])
    df['identifier'] = (df['Source Code'].astype(str) + df['Indicator Code'].astype(str) +  df['Sex Code'].astype(str) ).astype(int)
    
    for each in df['identifier'].unique():
        item_row = (df[df['identifier'] == each]).iloc[0]
        item_name =str(item_row['Source'])+ " " + str(item_row['Indicator'] + " " + str(item_row['Sex']))
        
#         if each == 210326125:
#             print("hello", i )
        
#         if each == '210326125':
#             print("hello", i )
            
        dict_meaning[each] = item_name
    
    
    df_step = df.filter(['Area Code', 'Year', 'identifier', 'Value'])
    
    df_final = df_step.pivot_table('Value', ['Year', 'Area Code'], 'identifier')
    
    return df_final
```

```python
def subset_least_missing(full_merged, n):

    percent_missing = full_merged.isnull().sum() * 100 / len(full_merged)
    missing_value_df = pd.DataFrame({'column_name': full_merged.columns,
                                     'percent_missing': percent_missing})

    missing_value_df.sort_values('percent_missing', inplace=True)
#     print("first", n)
#     average_missing = missing_value_df[:n]['percent_missing'].mean()
#     print("average percentage missing for each feature",average_missing )
    subset_final = missing_value_df.head(n)

    subset_final_merged = full_merged.filter(items =subset_final.column_name.tolist())
    print("new shape", subset_final_merged.shape)
    
    print("total percentage missing", np.count_nonzero(subset_final_merged.isna()) / subset_final_merged.size)
    # return the top least missing value    
    return subset_final_merged
```
